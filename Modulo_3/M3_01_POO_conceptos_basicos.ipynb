## Recordatorio rápido

En programación solemos combinar:

- **Datos**: lo que sabemos (estado).
- **Funciones**: lo que hacemos (comportamiento).

La POO agrupa ambos en una unidad: el **objeto**.

## Conceptos clave

- **Clase**: plantilla que define atributos y métodos.
- **Objeto (instancia)**: un elemento concreto creado a partir de la clase.
- **Atributo**: dato asociado al objeto o a la clase.
- **Método**: función definida dentro de la clase.

> Nota: en Python, todo es objeto, y las clases también lo son.

### `__init__` y `self`

- `__init__` se ejecuta al **crear** una instancia. Es una especie de constructor que obliga a que todos los objetos se creen siguiendo el mismo molde.
- `self` representa la instancia actual (convención).

```python
# __init__ sin parámetros: todos los objetos parten igual (constantes por defecto)
class Coche:
    def __init__(self):
        self.marca = "Desconocida"
        self.modelo = "Genérico"
        self.km = 0

    def conducir(self, distancia):
        self.km += distancia
        return self.km

    def descripcion(self):
        return f"{self.marca} {self.modelo} ({self.km} km)"

c1 = Coche()
c2 = Coche()

print(c1.descripcion())
print(c2.descripcion())

c1.conducir(100)
print(c1.descripcion())
print(c2.descripcion())
```

```python
# __init__ con parámetros: todos los objetos se crean con valores personalizados
class Coche:
    def __init__(self, marca, modelo, km=0):
        self.marca = marca
        self.modelo = modelo
        self.km = km

    def conducir(self, distancia):
        self.km += distancia
        return self.km

    def descripcion(self):
        return f"{self.marca} {self.modelo} ({self.km} km)"

mi_coche = Coche(marca="Toyota", modelo="Corolla", km=12_000)
print(mi_coche.descripcion())
print("KM tras conducir 150:", mi_coche.conducir(150))
print(mi_coche.descripcion())
```

## Atributos de instancia vs. atributos de clase

- **Instancia**: cada objeto tiene su propio valor.
- **Clase**: se comparte entre todas las instancias (salvo “sombreado” en una instancia).

Esto es muy útil para:
- constantes de configuración
- contadores globales
- valores por defecto compartidos

```python
class Alumno:
    centro = "Titan Academy"  # atributo de clase (compartido)

    def __init__(self, nombre, nota=0):
        self.nombre = nombre
        self.nota = nota

a1 = Alumno(nombre="Ana", nota=7.5)
a2 = Alumno(nombre="Luis", nota=9.0)

print(a1.centro, a2.centro)

Alumno.centro = "Titan 9001 School"
print(a1.centro, a2.centro)

# Sombreado (solo para una instancia)
a1.centro = "Centro Personalizado"
print("a1.centro:", a1.centro)
print("a2.centro:", a2.centro)
print("Alumno.centro:", Alumno.centro)
```

## Representación de un objeto como diccionario en Python

```python
# Mostrar como un objeto se puede visualizar como un diccionario
class Alumno:
    def __init__(self, nombre, nota=0):
        self.nombre = nombre
        self.nota = nota

a1 = Alumno(nombre="Ana", nota=7.5)
a2 = Alumno(nombre="Luis", nota=9.0)

print("a1: ", a1.__dict__)
print("a2: ", a2.__dict__)  
```

## Ejemplo guiado: cuenta bancaria

Objetivo: modelar una cuenta con invariantes simples:

- El saldo no debería volverse negativo.
- No tiene sentido ingresar/retirar cantidades <= 0.

La idea importante no es el “banco”, sino practicar:
- atributos
- validación
- métodos que modifican estado

```python
class CuentaBancaria:
    def __init__(self, titular, saldo=0.0):
        self.titular = titular
        self.saldo = float(saldo)

    def ingresar(self, cantidad):
        cantidad = float(cantidad)
        if cantidad <= 0:
            raise ValueError("La cantidad a ingresar debe ser positiva.")
        self.saldo += cantidad
        return self.saldo

    def retirar(self, cantidad):
        cantidad = float(cantidad)
        if cantidad <= 0:
            raise ValueError("La cantidad a retirar debe ser positiva.")
        if cantidad > self.saldo:
            raise ValueError("Saldo insuficiente.")
        self.saldo -= cantidad
        return self.saldo

    def __repr__(self):
        return f"CuentaBancaria(titular={self.titular!r}, saldo={self.saldo:.2f})"

c = CuentaBancaria(titular="Cristian", saldo=100)
print(c)
c.ingresar(50)
print(c)
c.retirar(30)
print(c)
```

## Trampas comunes y buenas prácticas

1) **Olvidar `self`** en la firma del método (error típico al empezar).  
2) Confundir atributo de clase con atributo de instancia.  
3) Escribir validación en muchos sitios: mejor centralizar (veremos `@property`).  
4) Exponer estado sin control cuando hay invariantes (saldo, edad, etc.).
