## ¿Qué es una función?

Una **función** es un bloque de código que realiza una tarea concreta y que podemos **reutilizar** varias veces en distintos puntos del programa.

Podemos verla como:
- Una **pequeña máquina** a la que le damos unos datos de entrada.
- La máquina realiza unos pasos.
- Al final nos devuelve (opcionalmente) un **resultado**.

Ventajas de usar funciones:
- Evitamos **repetir código**.
- Organizamos mejor el programa.
- Hacemos el código más fácil de leer y mantener.


## Definir y llamar a funciones

### Primera función

Para definir una función en Python usamos la palabra clave `def`, seguida del nombre de la función, paréntesis y dos puntos.  
El código que forma parte de la función va indentado (con sangría) debajo.


```python
def saludar():
    print("Hola desde la función saludar")

print("Inicio del programa")
saludar()
print("Final del programa")
```

### Funciones con parámetros

Una función puede recibir valores de entrada llamados **parámetros** (en la definición) o **argumentos** (cuando la llamamos).

Esto permite que la misma función haga su trabajo con datos distintos.


```python
def saludar_persona(nombre):
    print("Hola", nombre)


saludar_persona("Cristian")
saludar_persona("Laura")
```

### Parámetros posicionales y nombrados

Cuando llamamos a una función, podemos pasar los argumentos:
- **Por posición**: en el mismo orden en el que aparecen los parámetros.
- **Por nombre**: indicando explícitamente el nombre del parámetro.

Es importante entender que lo que se envía a la función es el **valor**, no la variable original.


```python
def presentar(nombre, edad):
    print("Me llamo", nombre, "y tengo", edad, "años")

# Argumentos por posición
presentar("Cristian", 30)

# Argumentos por nombre (el orden ya no importa)
presentar(edad=25, nombre="Ana")
presentar(nombre="Ana", edad=25)

e = 30
n = "Daniel"
presentar(nombre=n, edad=e)

nombre = "Sofía"
edad = 28
presentar(nombre=nombre, edad=edad)
```

Podemos comprobar que los nombres de las variables dentro de la función son independientes
de los nombres de las variables fuera de la función.


```python
def cambiar_valor(x):
    print("Dentro de la función, antes:", x)
    x = 99
    print("Dentro de la función, después:", x)


numero = 10
print("Fuera de la función, antes:", numero)
cambiar_valor(numero)
print("Fuera de la función, después:", numero)
```

### Parámetros con valores por defecto

Podemos asignar un **valor por defecto** a un parámetro.  
Si al llamar a la función no se indica ese argumento, se utilizará el valor por defecto.


```python
def saludar_formal(nombre, tratamiento="Sr."):
    print("Hola", tratamiento, nombre)


saludar_formal("García")
saludar_formal("López", tratamiento="Sra.")
```

## Devolver valores con `return`

Hasta ahora nuestras funciones solo mostraban información por pantalla.  
Sin embargo, muchas veces queremos que la función **devuelva un resultado** para poder seguir trabajando con él.

Para eso usamos la palabra clave `return`.


### `return` simple


```python
def sumar(a, b):
    resultado = a + b
    return resultado

resultado = sumar(3,5)
print("El resultado es:", resultado)
```

```python
# Esta función no devuelve nada, porque realiza varias acciones
# No es recomendable, es mejor que realice una unica tarea y devuelva el resultado
def sumar(a, b):
    resultado = a + b
    print("El resultado es:", resultado)

sumar(3, 5)
```

```python
# Importante: esta función no devuelve nada, por lo que el resultado es None
def sumar(a, b):
    resultado = a + b

resultado = sumar(3, 5)
print("El resultado es:", resultado)
```

### `return` múltiple


Una función puede devolver más de un valor a la vez.  
En realidad devuelve una **tupla**, pero podemos desempaquetarla en varias variables.


```python
def calcular_estadisticas(numeros):
    minimo = min(numeros)
    maximo = max(numeros)
    media = sum(numeros) / len(numeros)
    return minimo, maximo, media


lista_numeros = [10, 20, 30, 40]

minimo, maximo, media = calcular_estadisticas(lista_numeros)

print("Mínimo:", minimo)
print("Máximo:", maximo)
print("Media:", media)
```

## Ámbito de las variables

El **ámbito** de una variable indica desde dónde se puede usar esa variable.

- Una **variable local** se define dentro de una función y solo existe dentro de ella.
- Una **variable global** se define fuera de cualquier función y se puede usar desde el resto del archivo (con cuidado).


### Variables locales


```python
def ejemplo_local():
    mensaje = "Estoy dentro de la función"
    print(mensaje)


ejemplo_local()

# La siguiente línea daría error si la descomentamos, porque mensaje no existe fuera de la función:
# print(mensaje)
```

### Variables globales


```python
mensaje_global = "Estoy fuera de la función"


def mostrar_global():
    print(mensaje_global)


mostrar_global()
print(mensaje_global)
```

### Uso puntual de `global`

En general es mejor evitar modificar variables globales dentro de funciones, porque hace el código más difícil de seguir.

Python permite hacerlo con la palabra clave `global`, pero se recomienda usarla solo en casos muy concretos.


```python
contador = 0


def incrementar():
    global contador
    contador = contador + 1
    print("Dentro de la función, contador vale:", contador)


incrementar()
incrementar()
print("Fuera de la función, contador vale:", contador)
```

## `if __name__ == "__main__":` y función `main()`

Cuando escribimos programas en archivos `.py`, es habitual definir una función principal llamada `main()` 
que orquesta las llamadas al resto de funciones.

La condición `if __name__ == "__main__":` se utiliza para indicar que cierto código solo debe ejecutarse
cuando el archivo se ejecuta directamente, y no cuando se importa desde otro archivo.

En un notebook no se aprecia la diferencia, pero el patrón es el siguiente:


```python
def main():
    print("Este sería el punto de entrada principal del programa")
    # Aquí llamaríamos a otras funciones

if __name__ == "__main__":
    main()
```

## Diseñar programas con funciones

Una vez que conocemos las funciones, podemos organizar un programa de distintas formas.


### Opción 1: orquestar desde `main`


En esta estrategia, la función `main()` se encarga de coordinar los pasos del programa:

1. Llamar a una función para obtener datos.
2. Llamar a otra función para procesarlos.
3. Llamar a otra función para mostrar los resultados.

De este modo, `main()` tiene una visión clara del flujo general.


```python
def obtener_temperaturas():
    temperaturas = [18, 21, 19, 22]
    return temperaturas


def calcular_media(temperaturas):
    return sum(temperaturas) / len(temperaturas)


def mostrar_resultado(media):
    print("La temperatura media es:", media)


def main():
    datos = obtener_temperaturas()
    media = calcular_media(datos)
    mostrar_resultado(media)

if __name__ == "__main__":
    main()
```

### Opción 2: funciones que llaman a otras funciones


En esta estrategia, una función puede llamar a otra, y esta a su vez a otra.  
El valor devuelto con `return` va "subiendo" por la cadena de llamadas hasta llegar al origen.


```python
def obtener_precio_base():
    print("Obteniendo precio base")
    return 100


def aplicar_iva(precio):
    print("Aplicando IVA")
    return precio * 1.21


def aplicar_descuento(precio):
    print("Aplicando descuento")
    return precio * 0.9


def calcular_precio_final():
    precio_base = obtener_precio_base()
    precio_con_iva = aplicar_iva(precio_base)
    precio_final = aplicar_descuento(precio_con_iva)
    print("Precio final calculado dentro de la función:", precio_final)
    return precio_final


resultado = calcular_precio_final()
print("Resultado final recibido fuera de la función:", resultado)
```

En programas más grandes conviene combinar ambas ideas:

- Tener una función principal que orquesta las tareas.
- Dividir el trabajo en funciones más pequeñas que, si es necesario, llamen a otras funciones.

A continuación veremos un ejemplo.

```python
def leer_numero():
    texto = input("Introduce un número: ")
    numero = float(texto)
    return numero

def leer_varios_numeros(cuantos):
    numeros = []
    for i in range(cuantos):
        print("Número", i + 1)
        n = leer_numero()      # Aquí una función llama a otra función
        numeros.append(n)
    return numeros

def calcular_media(numeros):
    return sum(numeros) / len(numeros)

def main():
    print("Vamos a leer 3 números y calcular su media.\n")

    numeros = leer_varios_numeros(3)   # main llama a una función...
    print("\nHas introducido:", numeros)

    media = calcular_media(numeros)    # ...y luego a otra
    print("La media de los números es:", media)

if __name__ == "__main__":
    main()
```

## Argumentos indeterminados

A veces no sabemos de antemano cuántos argumentos vamos a necesitar.  
Python permite definir funciones que aceptan un número variable de argumentos.


### `*args` (número variable de argumentos posicionales)


```python
def sumar_todos(*numeros):
    total = 0
    for n in numeros:
        total = total + n
    print("La suma de todos los números es:", total)


sumar_todos(1, 2, 3)
sumar_todos(5, 10, 15, 20)
```

### `**kwargs` (número variable de argumentos nombrados)


```python
def mostrar_info(**datos):
    for clave, valor in datos.items():
        print(clave, "->", valor)


mostrar_info(nombre="Cristian", edad=30, ciudad="Madrid")
```

## Funciones anónimas (`lambda`)

Las funciones `lambda` son funciones pequeñas que se definen en una sola línea y no tienen un nombre propio.
Suelen usarse cuando necesitamos una función muy sencilla como argumento de otra función.

Su sintaxis básica es:

`lambda parámetros: expresión`


```python
# Función normal
def cuadrado(x):
    return x * x

# Función lambda equivalente
cuadrado_lambda = lambda x: x * x  # aunque se suele usar directamente, sin asignarla

print(cuadrado(5))
print(cuadrado_lambda(5))
```

```python
# Función normal
def calcular_iva_funcion(precio):
    return precio * 1.21

# Función lambda equivalente
calcular_iva_lambda = lambda precio: precio * 1.21

print("Precio con IVA de 100 (función normal):", calcular_iva_funcion(100))
print("Precio con IVA de 100 (lambda):", calcular_iva_lambda(100))
```

```python
# Función normal
def es_mayor_de_edad_funcion(edad):
    return edad >= 18

# Función lambda equivalente
es_mayor_de_edad_lambda = lambda edad: edad >= 18

print("20 es mayor de edad (función normal):", es_mayor_de_edad_funcion(20))
print("20 es mayor de edad (lambda):", es_mayor_de_edad_lambda(20))

print("15 es mayor de edad (función normal):", es_mayor_de_edad_funcion(15))
print("15 es mayor de edad (lambda):", es_mayor_de_edad_lambda(15))
```

```python
# Podemos añadir en lambda una condición con if-else

# Función normal con if / else
def es_mayor_de_edad_funcion(edad):
    if edad >= 18:
        return True
    else:
        return False

# Función lambda equivalente con if / else en una sola expresión
es_mayor_de_edad_lambda = lambda edad: True if edad >= 18 else False

print("20 es mayor de edad (función normal):", es_mayor_de_edad_funcion(20))
print("20 es mayor de edad (lambda):", es_mayor_de_edad_lambda(20))

print("15 es mayor de edad (función normal):", es_mayor_de_edad_funcion(15))
print("15 es mayor de edad (lambda):", es_mayor_de_edad_lambda(15))
```

## Documentar funciones (docstrings)

Es una buena práctica añadir una pequeña descripción al inicio de cada función para explicar qué hace,
qué parámetros recibe y qué devuelve.

Esta descripción se escribe como una cadena de texto entre comillas triples justo debajo de la definición.


```python
def sumar(a, b):
    """Devuelve la suma de dos números a y b."""
    return a + b


resultado = sumar(2, 3)
print("Resultado de sumar:", resultado)
```

### Docstrings ampliados y *type hints* (anotaciones de tipo)

En algunas funciones vas a ver un docstring más completo que incluye apartados como `Args:` y `Returns:`.  
Este estilo no forma parte de la sintaxis oficial de Python, sino que es una **convención muy extendida en la comunidad**, especialmente el conocido como *estilo Google*.  

Python únicamente exige (según la PEP 257) que el docstring:
- esté entre comillas triples,
- y que la primera línea describa brevemente qué hace la función.

Todo lo demás —como añadir secciones estructuradas para los parámetros o el valor de retorno— es opcional y forma parte de buenas prácticas de documentación, muy útiles en proyectos grandes o cuando se genera documentación automática.

Por otro lado, también pueden aparecer anotaciones como `a: float` o `-> float`.  
Estas son **anotaciones de tipo**, conocidas como *type hints*. Su objetivo es indicar qué tipo de dato se espera en cada parámetro y cuál será el tipo del valor devuelto.  

Es importante saber que:
- No son obligatorias.
- Python **no las usa** para validar tipos en tiempo de ejecución: son solo “pistas” para quien lee el código y para herramientas de ayuda (editores, linters, analizadores estáticos).
- Ayudan a que el código sea más claro y fácil de mantener.

En resumen:  
- Los docstrings ampliados ayudan a documentar mejor las funciones.  
- Los *type hints* ayudan a entender qué tipos espera usar la función.  
Ambos son elementos modernos y muy recomendables, pero opcionales.


```python
def sumar(a: float, b: float) -> float:
    """Devuelve la suma de dos números.

    Args:
        a: Primer número.
        b: Segundo número.

    Returns:
        La suma de a y b.
    """
    return a + b
```

## Notas y buenas prácticas

- Intenta que cada función tenga una **responsabilidad clara**.
- Evita que una función haga demasiadas cosas distintas.
- Prefiere pasar datos como parámetros y devolver resultados con `return`,
  en lugar de depender de muchas variables globales.
- Utiliza nombres descriptivos para las funciones y los parámetros.
- No tengas miedo de crear muchas funciones pequeñas si eso hace el código más claro.
