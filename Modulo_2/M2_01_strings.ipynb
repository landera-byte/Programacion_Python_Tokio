## Recordatorio rápido

Un **string** (`str`) es una secuencia de caracteres Unicode.  
Se puede definir con comillas simples `'...'`, dobles `"..."` o triples `'''...'''` / `"""..."""` (multilínea).

Podemos:
- Acceder a una posición con `s[i]`
- Hacer slicing con `s[inicio:fin:paso]`


```python
s = "Python"

print(s)       # Python
print(s[0])    # P
print(s[-1])   # n

print(s[0:3])  # Pyt
print(s[:3])   # Pyt (equivale a 0:3)
print(s[3:])   # hon
print(s[::2])  # Pto (paso 2)
```

## Conceptos clave

### Inmutabilidad
Los `str` son **inmutables**: no puedes modificar un carácter "en el sitio".  
En su lugar, creas un **nuevo** string.

### Longitud, pertenencia y recorrido
- `len(s)` devuelve la longitud.
- `'sub' in s` comprueba si una subcadena aparece.
- Se puede recorrer con `for`.


```python
s = "hola"
print(len(s))         # 4
print("ol" in s)      # True

# Inmutabilidad:
# s[0] = "H"  # TypeError

s2 = "H" + s[1:]
print(s2)             # Hola

for ch in "abc":
    print(ch)
```

## Literales, comillas y caracteres especiales

- Secuencias de escape típicas: `\n` (salto de línea), `\t` (tab), `\"` (comillas), `\\` (barra invertida).
- Strings "raw" (crudos) con prefijo `r"..."` son útiles para rutas en Windows y expresiones regulares
  (ojo: no pueden terminar en una sola `\`).


```python
print("Línea 1\nLínea 2")
print("Col1\tCol2")

ruta = r"C:\Users\cristian\proyecto"
print(ruta)
```

## Concatenación, repetición y rendimiento

Operaciones habituales:
- Concatenar: `a + b`
- Repetir: `'ha' * 3`
- Construir texto desde piezas: **mejor** `''.join(lista_de_piezas)` que concatenar en un bucle.


```python
print("Py" + "thon")
print("ha" * 3)

piezas = ["Esto", "se", "une", "bien"]
texto = " ".join(piezas)
print(texto)
```

## Búsqueda y extracción

### `find` vs `index`
- `s.find(sub)` devuelve el índice o `-1` si no lo encuentra.
- `s.index(sub)` devuelve el índice o lanza `ValueError` si no lo encuentra.

### `count`, `startswith`, `endswith`
Útiles para conteos y validaciones.

### `partition` / `rpartition`
Parten el string en 3 partes: (antes, separador, después).  
Si el separador no existe, devuelve (string_original, '', '').


```python
s = "curso_python_strings"

print(s.find("python"))      # 6
print(s.find("java"))        # -1

print(s.count("_"))          # 2
print(s.startswith("curso")) # True
print(s.endswith("s"))       # True

print(s.partition("_"))      # ('curso', '_', 'python_strings')
print(s.rpartition("_"))     # ('curso_python', '_', 'strings')
```

## Limpieza: `strip`, `replace`, `removeprefix/removesuffix`

### `strip`, `lstrip`, `rstrip`
Eliminan **caracteres** al principio y/o final (por defecto, espacios).  
Importante: `strip("abc")` NO elimina la subcadena `"abc"`; elimina cualquier combinación
de los caracteres `'a'`, `'b'`, `'c'` en los extremos.

### `removeprefix` / `removesuffix` (Python 3.9+)
Eliminan un prefijo/sufijo **exacto** si está presente.

### `replace`
Reemplaza subcadenas. Puedes limitar cuántas sustituciones con el tercer parámetro.


```python
texto = "   hola mundo   "
print(texto.strip())   # "hola mundo"

# Peligro común:
x = "abcXYZcba"
print(x.strip("abc"))  # "XYZ" (quita caracteres a/b/c en extremos)

url = "https://example.com"
print(url.removeprefix("https://"))  # "example.com"
print(url.removesuffix(".com"))      # "https://example"

s = "uno, dos, tres, dos"
print(s.replace("dos", "DOS", 1))    # solo la primera
```

## Tablas de traducción: `translate` y `maketrans`

`translate` permite reemplazar/eliminar muchos caracteres de una vez (muy rápido).
Se suele construir la tabla con `str.maketrans`.

Casos típicos:
- quitar signos de puntuación
- normalizar separadores
- traducir caracteres (p.ej., reemplazar acentos de forma controlada)


```python
texto = "¡Hola! ¿Qué tal? 100% seguro."

# Eliminar signos: asignar None al carácter
tabla = str.maketrans({
    "¡": None, "¿": None, "!": None, "?": None, "%": None, ".": None
})

print(texto.translate(tabla))
```

## Separar y unir

### `split` / `rsplit`
- `split()` sin argumento separa por cualquier whitespace consecutivo.
- `split(sep)` separa por un separador exacto.
- `maxsplit` limita el número de cortes.

### `splitlines`
Separa por saltos de línea de forma robusta.

### `join`
Une elementos iterables (normalmente una lista) en un string.


```python
s = "  uno   dos\tres\ncuatro  "
print(s.split())                # ['uno', 'dos', 'tres', 'cuatro']

csv = "a,b,c,d"
print(csv.split(","))           # ['a', 'b', 'c', 'd']
print(csv.split(",", maxsplit=1))  # ['a', 'b,c,d']

multilinea = "L1\nL2\r\nL3"
print(multilinea.splitlines())  # ['L1', 'L2', 'L3']

partes = ["2025", "12", "18"]
print("-".join(partes))         # 2025-12-18
```

## Mayúsculas/minúsculas y comparaciones seguras

Métodos comunes:
- `lower()`, `upper()`
- `capitalize()` (solo primera letra)
- `title()` (título, ojo con apóstrofes/idiomas)
- `swapcase()` (intercambia)
- `casefold()` (comparación *más agresiva* y recomendada para texto humano)

Para comparar texto "sin importar mayúsculas", usa `casefold()`:


```python
a = "Straße"      # La palabra calle en alemán puede contener el siguiente caracter: ß
b = "STRASSE"

print(a.lower() == b.lower())       # False (depende)
print(a.casefold() == b.casefold()) # True
```

## Validación de contenido: métodos `is*`

Algunos muy útiles:
- `isalnum()`, `isalpha()`, `isdigit()`, `isdecimal()`, `isnumeric()`
- `isspace()`, `islower()`, `isupper()`, `istitle()`
- `isascii()` (comprueba que todos los caracteres están en ASCII)

Ojo: estos métodos dependen de Unicode, no solo de "a-z".


```python
t = "123"  # Prueba con otras: "12.3", "café", "  ", "ABC", "Español", "hello!"

print(t, "->",
      "isdigit:", t.isdigit(),
      "isalpha:", t.isalpha(),
      "isalnum:", t.isalnum(),
      "isspace:", t.isspace(),
      "isascii:", t.isascii())
```

## Relleno y alineación

- `ljust`, `rjust`, `center` añaden relleno (por defecto espacios) hasta un ancho.
- `zfill` rellena con ceros a la izquierda (típico para códigos numéricos).


```python
codigo = "42"
print(codigo.zfill(5))      # 00042

titulo = "Python"
print(titulo.center(20, "-"))
print(titulo.ljust(10, "."))
print(titulo.rjust(10, "."))
```

## Formateo de texto moderno

### `f-strings` (recomendado)
Permiten interpolar expresiones con `{...}`:
- Rápidas
- Legibles
- Soportan el mini-lenguaje de formato (`:.2f`, `:>10`, `:%`, etc.)

> Nota moderna: en Python 3.12 se flexibiliza la sintaxis de f-strings (PEP 701),
> permitiendo expresiones más generales dentro de `{...}`.

### `str.format`
Sigue siendo útil, sobre todo en plantillas más genéricas.


```python
nombre = "Cristian"
edad = 30
pi = 3.14159265

print(f"Hola {nombre}, tienes {edad} años.")
print(f"pi con 3 decimales: {pi:.3f}")

# Alineación y ancho:
for n in [1, 20, 300]:
    print(f"{n:>5} -> {n:0>5}")  # derecha y con ceros

# self-documenting (Python 3.8+)
x = 10
y = 3
print(f"{x=}, {y=}, {x/y=:.2f}")

# str.format()
print("Hola {}, tienes {} años".format(nombre, edad))
```

### Mini-lenguaje de formato (algunos ejemplos)

- Números:
  - `:.2f` -> 2 decimales
  - `:,` -> separador de miles
  - `:%` -> porcentaje
- Alineación:
  - `:>10` derecha, ancho 10
  - `:<10` izquierda, ancho 10
  - `:^10` centrado, ancho 10


```python
importe = 1234567.891
ratio = 0.1234

print(f"{importe:,.2f}")  # 1,234,567.89 (según locale no cambia; usa coma fija)
print(f"{ratio:.1%}")     # 12.3%

palabra = "txt"
print(f"|{palabra:>10}|")
print(f"|{palabra:<10}|")
print(f"|{palabra:^10}|")
```

## Utilidades del módulo `string`

El módulo estándar `string` incluye constantes útiles (alfabeto, dígitos, puntuación…) y una clase `Template`
para plantillas sencillas.

- Constantes típicas: `string.ascii_letters`, `string.digits`, `string.punctuation`, `string.whitespace`
- Plantillas: `string.Template` (sustitución con `$nombre`)


```python
import string
from string import Template

print(string.ascii_letters)
print(string.digits)
print(string.punctuation)

tpl = Template("Hola $nombre, tu pedido $id está listo.")
print(tpl.substitute(nombre="Cristian", id=123))
```

## Texto multilínea: `textwrap` y utilidades

Cuando generas texto (logs, informes, prompts, emails…), suele ser útil:
- Ajustar el ancho: `textwrap.fill`
- Quitar sangría común: `textwrap.dedent`
- Indentar: `textwrap.indent`


```python
import textwrap

parrafo = """Python es un lenguaje muy usado para automatización, análisis de datos y desarrollo.
Cuando generas texto largo, puede interesarte ajustarlo a un ancho fijo.
"""

print(textwrap.fill(parrafo, width=50))

bloque = """        línea 1
            línea 2
        línea 3
"""
print("DEDENT:")
print(textwrap.dedent(bloque))

print("INDENT:")
print(textwrap.indent("línea A\nlínea B", prefix="> "))
```

## Expresiones regulares (`re`) para patrones

Cuando necesitas buscar patrones complejos, `re` es la herramienta estándar.

Ideas típicas:
- Validar formatos (email, teléfono, DNI, etc.)
- Extraer datos
- Reemplazos avanzados

Consejos:
- Usa raw strings: `r"..."`.
- Compila patrones si vas a reutilizarlos (`re.compile`).


```python
import re

texto = "Pedidos: A-102, B-7, C-999."
pat = re.compile(r"[A-Z]-\d+")
print(pat.findall(texto))   # ['A-102', 'B-7', 'C-999']

# Capturas:
m = re.search(r"([A-Z])-(\d+)", "B-7")
if m:
    letra, numero = m.group(1), int(m.group(2))
    print(letra, numero)

# Reemplazo:
print(re.sub(r"\d+", "#", texto))
```

## Unicode y normalización (`unicodedata`)

Dos strings pueden "verse igual" pero ser diferentes a nivel Unicode (composición vs descomposición).  
Si haces comparaciones, búsquedas o deduplicación, a veces conviene normalizar.

`unicodedata.normalize(form, s)` con `form` en {"NFC", "NFD", "NFKC", "NFKD"}.
- **NFC**: composición canónica (muy común)
- **NFKC**: compatibilidad (más agresivo)


```python
import unicodedata

# 'é' puede ser un solo carácter o 'e' + acento combinante
s1 = "café"
s2 = "cafe\u0301"  # e + acento combinante

print(s1 == s2)  # a menudo False
print(unicodedata.normalize("NFC", s1) == unicodedata.normalize("NFC", s2))  # True
```

## `str` vs `bytes`: codificar y decodificar

- `str` = texto (Unicode)
- `bytes` = datos binarios

Para pasar de `str` a `bytes`: `encode(encoding)`  
Para pasar de `bytes` a `str`: `decode(encoding)`

UTF-8 es el estándar de facto.


```python
texto = "mañana ☀️"

b = texto.encode("utf-8")
print(b)                 # bytes
print(type(b))

t2 = b.decode("utf-8")
print(t2)
print(type(t2))
```

## Trampas comunes y buenas prácticas

- **No concatenes** en un bucle con `+=` si son muchas piezas → usa `''.join(...)`.
- Para comparar sin mayúsculas y de forma robusta en Unicode → `casefold()`.
- Si limpias prefijos/sufijos exactos → `removeprefix/removesuffix` (mejor que `strip`).
- Si tu pipeline hace "limpieza de texto", suele tener sentido:
  1) `strip` (espacios)
  2) normalización Unicode (`unicodedata.normalize`)
  3) minúsculas (`casefold`)
  4) reemplazos/`translate`
  5) tokenización (`split`, regex, etc.)
