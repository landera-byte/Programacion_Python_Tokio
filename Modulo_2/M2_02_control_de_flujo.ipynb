## ¿Qué es el control de flujo?

Hasta ahora hemos visto cómo guardar datos y hacer operaciones con ellos.  
El **control de flujo** es lo que decide **qué instrucciones se ejecutan y cuántas veces**.

Gracias al control de flujo podemos:
- Ejecutar un bloque de código **solo si** se cumple una condición (condicionales).
- Repetir un bloque de código varias veces (bucles).


## Condicionales

### ¿Qué es una condición?

Una **condición** es una expresión que se puede evaluar como `True` o `False`.  
Se suele construir usando operadores de comparación (`>`, `<`, `==`, etc.) y lógicos (`and`, `or`, `not`).

Los condicionales permiten que el programa tome decisiones en función de esas condiciones.


### Sentencia `if`


```python
edad = 20

if edad >= 18:
    print("Eres mayor de edad")
```

### `if` + `else`


```python
edad = 15

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

### `if` + `elif` + `else`


```python
nota = 7

if nota >= 9:
    print("Sobresaliente")
elif nota >= 7:
    print("Notable")
elif nota >= 5:
    print("Aprobado")
else:
    print("Suspenso")
```

```python
# Mejoremos lo anterior:

nota = float(input("Introduce la nota del 0 al 10: "))

if nota >= 9 and nota <= 10:
    print("Sobresaliente")
elif nota >= 7 and nota < 9:
    print("Notable")
elif nota >= 6 and nota < 7:
    print("Aprobado")
elif nota >= 5 and nota < 6:
    print("Suficiente")
elif nota >= 0 and nota < 5:
    print("Suspenso")
else:
    print("La nota no es válida (debe estar entre 0 y 10)")
```

### Condicionales anidados


```python
pais = "España"
edad = 20

if pais == "España":
    if edad >= 18:
        print("En España eres mayor de edad")
    else:
        print("En España aún eres menor de edad")
else:
    print("No se ha definido la mayoría de edad para este país")
```

### Expresión condicional (forma compacta)

A veces queremos elegir entre dos valores de forma muy corta.  
Python permite escribir una **expresión condicional** en una sola línea:

`valor_si_true if condicion else valor_si_false`

Es útil para asignar valores simples sin escribir un bloque `if` completo.

Este método no es compatible con una estructura if-elif-else


```python
edad = 20

mensaje = "Mayor de edad" if edad >= 18 else "Menor de edad"
print(mensaje)
```

### `match` / `case` (Python 3.10+)

En versiones modernas de Python (3.10 en adelante) existe la sentencia `match`, 
que permite comparar un valor contra varios posibles casos de forma más clara.

Se parece a un `switch` de otros lenguajes, pero es más potente.


```python
dia = "lunes"

match dia:
    case "lunes":
        print("Empezamos la semana")
    case "viernes":
        print("Casi fin de semana")
    case "sábado" | "domingo":
        print("Es fin de semana")
    case _:
        print("Día normal")
```

## Bucles

Los bucles permiten **repetir** un bloque de código varias veces.  
En Python los dos bucles principales son:

- `while`: repite mientras se cumpla una condición.
- `for`: recorre una secuencia de valores (por ejemplo, una lista o un rango de números).


### Bucle `while`


```python
contador = 1

while contador <= 5:
    print("Contador vale:", contador)
    contador = contador + 1

print("Fin del bucle while")
```

Es importante asegurarse de que la condición del `while` **cambie en algún momento**, 
para que el bucle no se ejecute para siempre.

A continuación un ejemplo de bucle infinito. __CUIDADO SI LO EJECUTAS. PARALO RAPIDAMENTE PARA NO SATURAR EL FICHERO__


```python
contador = 0
while True:
    contador += 1
    print(contador)
```

### Bucle `for`


```python
for numero in [1, 2, 3, 4, 5]:
    print("Número:", numero)

print("Fin del bucle for")
```

```python
nombres = ["Cristian", "Eva", "Sara"]
for n in nombres:
    print("Nombre:", n)

print("Fin del bucle for")
```

### Usar `enumerate()` con `for`

```python
frutas = ["manzana", "pera", "plátano", "kiwi"]

for indice, fruta in enumerate(frutas):
    print("Índice:", indice, "→ Fruta:", fruta)

# Empezar la numeración en 1 en vez de 0
for numero, fruta in enumerate(frutas, start=1):
    print("Número:", numero, "→ Fruta:", fruta)
```

### Usar `range()` con `for`


```python
# range(inicio, fin) genera números desde inicio hasta fin-1
for i in range(1, 6):
    print("Iteración:", i)
```

```python
# range(inicio, fin, salto)

print("De 0 a 10 de 2 en 2:")
for i in range(0, 11, 2):
    print(i)

print("\nDe 10 a 0 hacia atrás:")
for i in range(10, -1, -1):
    print(i)

print("\nContar de 5 en 5:")
for i in range(0, 31, 5):
    print(i)
```

### Instrucciones `break` y `continue`


```python
# break: sale del bucle
for numero in range(1, 10):
    if numero == 5:
        break
    print("Número con break:", numero)

# continue: salta a la siguiente iteración
for numero in range(1, 10):
    if numero % 2 == 0:
        continue
    print("Número impar con continue:", numero)
```

## Recorrer colecciones con bucles

Una de las formas más habituales de usar bucles es para recorrer colecciones 
como listas, tuplas, conjuntos y diccionarios.


### Recorrer una lista


```python
frutas = ["manzana", "pera", "plátano"]

for fruta in frutas:
    print("Fruta:", fruta)
```

### Recorrer una tupla


```python
puntos = (10, 20, 30)

for punto in puntos:
    print("Puntuación:", punto)
```

### Recorrer un conjunto (`set`)


```python
conjunto = {"rojo", "verde", "azul"}

for color in conjunto:
    print("Color:", color)
```

### Recorrer un diccionario


```python
persona = {"nombre": "Cristian", "edad": 30, "pais": "España"}

# Recorrer solo las claves
for clave in persona.keys():
    print("Clave:", clave)

# Recorrer claves y valores a la vez
for clave, valor in persona.items():
    print(clave, "->", valor)
```

## Ejemplos combinando bucles y condicionales

En la práctica, es muy común combinar bucles y condicionales para filtrar, buscar o procesar datos.


### Ejemplo: filtrar países por población mínima


```python
paises = [
    {"nombre": "España", "poblacion_millones": 47},
    {"nombre": "Portugal", "poblacion_millones": 10},
    {"nombre": "Francia", "poblacion_millones": 67},
    {"nombre": "Andorra", "poblacion_millones": 0.08},
]

poblacion_minima = 20

for pais in paises:
    if pais["poblacion_millones"] >= poblacion_minima:
        print(pais["nombre"], "tiene más de", poblacion_minima, "millones de habitantes")
```

### Ejemplo: buscar un país concreto


```python
paises = ["España", "Portugal", "Francia", "Italia"]
buscado = "Francia"

encontrado = False

for pais in paises:
    if pais == buscado:
        encontrado = True
        break

if encontrado:
    print("Se ha encontrado el país:", buscado)
else:
    print("No se ha encontrado el país:", buscado)
```

## Notas modernas y recordatorios

- La sentencia `match` / `case` está disponible a partir de Python 3.10 y permite 
  escribir estructuras de decisión más claras cuando hay muchos casos posibles.
- La expresión condicional (`valor_si_true if condicion else valor_si_false`) 
  puede simplificar código cuando solo necesitamos elegir entre dos opciones.
- `for` se usa con mucha frecuencia para recorrer colecciones (listas, tuplas, conjuntos, diccionarios),
  por lo que es muy importante practicarlo junto con las estructuras de datos del módulo anterior.
