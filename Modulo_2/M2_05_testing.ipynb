## 1) ¿Qué son las pruebas de software?

Un **test** es código que verifica que **otro código** se comporta como esperamos.

En este módulo nos centramos en **tests unitarios**:
- Prueban una función o componente pequeño
- Son rápidos
- No dependen de red, bases de datos, APIs reales, etc. (en general)

(Después existen pruebas de integración, end‑to‑end, etc., pero para empezar lo importante es lo unitario.)


## 2) ¿Por qué hacer tests?

Los tests sirven para:

- **Detectar errores antes** (y más barato).
- Evitar “romper” cosas al cambiar código (**regresión**).
- Documentar con ejemplos qué debe hacer una función.
- Ganar confianza para refactorizar.

Idea clave:
> Un proyecto con tests permite cambiar código con menos miedo.


## 3) Herramientas: `unittest` vs `pytest`

- **`unittest`**: está en la biblioteca estándar de Python (no hay que instalar nada) y se usa aún, sobre todo en código legado.
- **`pytest`**: es la opción más extendida hoy en proyectos modernos porque es muy cómoda:
  - tests como funciones sencillas
  - `assert` “normal”
  - buena salida de errores
  - fixtures y plugins (cuando los necesites)

En este módulo usaremos **`pytest`** por ser lo más habitual en la práctica.

## 5) Ejemplo 1: primer test con `assert`

### `ejemplo1.py`
Crea el fichero `ejemplo1.py` con este contenido:

````python
def suma(a, b):
    return a + b
````

### `test_ejemplo1.py`
Crea el fichero `test_ejemplo1.py` con:


````python
from ejemplo1 import suma

def test_suma_basica():
    assert suma(2, 3) == 5

def test_suma_con_negativos():
    assert suma(-2, 10) == 8
````

## Cómo ejecutar las pruebas (pytest)

Asegúrate de que en la misma carpeta tienes estos dos archivos:

- `ejemplo1.py`
- `test_ejemplo1.py`

Instala `pytest` (solo la primera vez):

```bash
pip install pytest
```

Ejecuta las pruebas desde esa carpeta:

- Ejecutar **todas** las pruebas que encuentre:

```bash
pytest
```
o
```bash
python -m pytest
```

- Ejecutar **solo** este fichero de tests:

```bash
pytest test_ejemplo1.py
```
o

```bash
python -m pytest test_ejemplo1.py
```

- Ver salida más detallada:

```bash
pytest -v
```
o
```bash
pytest -v test_ejemplo1.py
```
o

```bash
python -m pytest -v test_ejemplo1.py
```

Si todo está bien, verás algo como `2 passed`.

## 6) Ejemplo 2: testear excepciones con `pytest.raises`

### `ejemplo2.py`

````python
def dividir(a, b):
    if b == 0:
        raise ValueError("No se puede dividir entre cero")
    return a / b
````

### `test_ejemplo2.py`

````python
import pytest
from ejemplo2 import dividir

def test_dividir_ok():
    assert dividir(10, 2) == 5

def test_dividir_por_cero_lanza_error():
    with pytest.raises(ValueError):
        dividir(10, 0)
````

**Idea importante:** también es correcto que una función *falle* si el input es inválido…
siempre que falle **de forma controlada** (lanzando una excepción adecuada).

## 7) Ejemplo 3: varios casos (parametrización)

Cuando quieres probar muchos casos sin repetir código, `pytest` permite **parametrizar**.

### `ejemplo3.py`

````python
def es_par(n: int) -> bool:
    return n % 2 == 0
````

### `test_ejemplo3.py`
Parametrizar = “una misma prueba, con muchas entradas”.

````python
import pytest
from ejemplo3 import es_par

@pytest.mark.parametrize(
    "n, esperado",
    [
        (0, True),
        (1, False),
        (2, True),
        (11, False),
        (-4, True),
    ],
)
def test_es_par(n, esperado):
    assert es_par(n) is esperado
````

## 8) Caso final: la pizzería (excepciones propias + tests)

Aquí juntamos:
- Validación de entradas
- Excepciones propias (como en el módulo de excepciones)
- Tests que verifican comportamiento correcto e incorrecto


### `pizza.py`
Crea `pizza.py`:

````python
class PizzaError(Exception):
    """Excepción base para errores relacionados con pizzas."""

class TipoPizzaInvalido(PizzaError):
    """Se pidió un tipo de pizza que no existe en el sistema."""

class TamanoPizzaInvalido(PizzaError):
    """Tamaño no permitido para ese tipo de pizza."""


MENU = {
    "margarita": {"pequena", "mediana", "familiar"},
    "barbacoa": {"pequena", "mediana", "familiar"},
    "cuatro quesos": {"mediana"},
}

def crear_pizza(tipo: str, tamano: str) -> str:
    tipo = tipo.strip().casefold()
    tamano = tamano.strip().casefold()

    if tipo not in MENU:
        raise TipoPizzaInvalido(f"Tipo inválido: {tipo!r}. Disponibles: {sorted(MENU)}")

    if tamano not in MENU[tipo]:
        raise TamanoPizzaInvalido(
            f"Tamaño inválido {tamano!r} para {tipo!r}. Permitidos: {sorted(MENU[tipo])}"
        )

    return f"Pizza creada: {tipo} ({tamano})"
````

### `test_pizza.py`
Crea `test_pizza.py`:

````python
import pytest
from pizza import crear_pizza, TipoPizzaInvalido, TamanoPizzaInvalido

def test_crear_pizza_ok():
    assert crear_pizza("margarita", "mediana") == "Pizza creada: margarita (mediana)"

def test_tipo_invalido():
    with pytest.raises(TipoPizzaInvalido):
        crear_pizza("carbonara", "mediana")

def test_tamano_invalido():
    with pytest.raises(TamanoPizzaInvalido):
        crear_pizza("cuatro quesos", "familiar")

@pytest.mark.parametrize(
    "tipo, tamano",
    [
        ("barbacoa", "gigante"),
        ("cuatro quesos", "pequena"),
    ],
)
def test_varios_tamanos_invalidos(tipo, tamano):
    with pytest.raises(TamanoPizzaInvalido):
        crear_pizza(tipo, tamano)
````

**¿Qué estamos validando?**
- Que para inputs correctos obtenemos una salida correcta.
- Que para inputs incorrectos, el sistema falla **con la excepción esperada**.

## 9) Chuleta de comandos `pytest`

Desde la carpeta `m2_05_testing/`:

```bash
pytest
```

Opciones útiles (para empezar):
- `pytest -q` → salida más corta
- `pytest -vv` → más detalle
- `pytest -x` → para en el primer fallo
- `pytest -k palabra` → ejecuta solo tests cuyo nombre contenga esa palabra

Ejemplo:
```bash
pytest -q
pytest -k pizza -vv
```

### Dudas típicas de principiantes

- **“¿Por qué mis tests no se ejecutan?”**  
  Asegúrate de que el fichero se llame `test_algo.py` o `algo_test.py` y que las funciones empiecen por `test_`.

- **“¿Dónde pongo el código?”**  
  Código “normal” en `ejemploX.py`. Tests en `test_ejemploX.py`.

- **“¿Puedo usar prints?”**  
  Sí, pero en tests es mejor usar `assert`. Los prints no verifican nada por sí solos.
